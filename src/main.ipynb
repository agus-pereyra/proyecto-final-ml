```python
import scipy.io as sio
import numpy as np
from pathlib import Path

data_path = Path(__file__).parent.parent / 'data' if '__file__' in dir() else Path('../data')

mat = sio.loadmat(data_path / 'Bislab01' / '1' / 'labels.mat')

# Aplanar de (1, 860) a (860,)
dreem_labels  = mat['dreem_label'].flatten()   # shape: (860,)
expert_labels = mat['expert_label'].flatten()  # shape: (860,)
rec_start     = mat['recStart'][0]             # '2021-12-10 22:35:53'

print(f"Total epochs: {len(expert_labels)}")
print(f"Duración: {len(expert_labels) * 30 / 3600:.1f} horas")

# Distribución de clases
nombres = {0: 'Wake', 1: 'N1', 2: 'N2', 3: 'N3', 4: 'REM'}
for clase, nombre in nombres.items():
    n = np.sum(expert_labels == clase)
    pct = n / len(expert_labels) * 100
    print(f"{nombre}: {n} epochs ({pct:.1f}%)")
```
